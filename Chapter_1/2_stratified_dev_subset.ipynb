#### In Gen AI it is not Training or Test Set, it is Development Set and Evaluation Set
- "Training set" / "test set" applies when you're fitting parameters to data — like training a TF-IDF + Logistic Regression classifier, or fine-tuning a model, where the model literally learns from labeled examples.

##### What we built (300 dev + 2,700 eval) is different. 
- It assumes you're not fitting parameters on this data — instead you're evaluating something like an LLM prompt, a zero-shot classifier, or a rule-based system, where:
- Dev set (300) = where you try different prompts/rules and see what works
- Eval set (2,700) = where you measure the final, honest performance — untouched during tuning

##### If that's your setup, there's no "training set" at all, because nothing is being trained.

```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv("../data/fd_dataset_messy.csv")

dev_set, eval_set = train_test_split(data, train_size = 300, stratify=data['label'], random_state=42)
print("Development Set Shape:", len(dev_set))
print("Evaluation Set Shape:", len(eval_set))
```

```python
# Writing to files
dev_set.to_csv('../data/dev_subset_300.csv', index=False)
eval_set.to_csv('../data/eval_set_2200.csv', index=False)
```
