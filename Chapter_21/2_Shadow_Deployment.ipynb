## Part A: Theory

### 1. Concept, Intuition, and Why This Exists

- Topic 1 exposed this project's pipeline behind a genuine, real, running API endpoint — currently serving Chapter 1's rule-based classifier directly, per that notebook's own actual implementation. But this project's real, complete architecture (built across Chapters 2-20) includes far more sophisticated GenAI classification, retrieval, and tool-calling capability that Topic 1's simple endpoint didn't yet incorporate. This topic addresses the natural, necessary question that gap raises: how does a project safely introduce a more sophisticated, but not-yet-fully-production-validated capability, without risking real customer impact if that capability has an undiscovered flaw?
- **Shadow deployment**, precisely: running a candidate system (this project's full GenAI pipeline) *silently, in parallel* with the currently-live, trusted system (Chapter 1's rule-based classifier) for every real, incoming request — the shadow system's output is logged and compared against the live system's output, but the shadow system's result is never actually used to determine the real, customer-facing response. This is a genuinely different risk profile from simply deploying the new system directly: any flaw in the shadow system, however severe, has zero actual customer impact, since its output is never used for anything beyond comparison and analysis.
- Why this matters specifically for this project's own real, demonstrated history of discovering flaws through careful evaluation: Chapter 15 Topic 1's own concrete demonstration (a flawed tool implementation passing every test against clean data, only failing on a realistic typo) is exactly the kind of subtle, hard-to-predict issue shadow deployment is designed to surface — running the full GenAI pipeline against genuine, real, live production traffic (not just Chapter 15's evaluation-set data) for an extended period, before ever trusting its output for actual customer-facing decisions, provides exactly the kind of real-world validation evidence this notebook series has repeatedly required before adopting any consequential change.


### 2. Internal Working — Step by Step

**Implementing shadow deployment for this project's actual pipeline, step by step:**

1. **For every real, incoming request Topic 1's endpoint receives, run both the live system (Chapter 1's rule-based classifier, currently serving real customer responses) and the shadow system (this project's full GenAI pipeline) on the exact same input.**
2. **Return only the live system's output as the actual API response** — the shadow system's processing happens genuinely in parallel, but its result is never surfaced to whatever called this project's API, ensuring zero customer-facing risk regardless of what the shadow system produces.
3. **Log both systems' outputs, tagged with the same request/trace identifier**, directly reusing Chapter 20 Topic 1's completed tracing infrastructure — this comparison data is precisely what makes shadow deployment valuable, not just a parallel-execution exercise with no actual analysis attached to it.
4. **Aggregate the comparison data over a genuine, extended period of real production traffic**, computing agreement rate (how often the shadow and live systems produce the same result) and specifically investigating disagreement cases — exactly the same three-way-partition methodology Chapter 16 Topic 2 and Chapter 18 Topic 7 both already applied to comparing two different systems, now applied to comparing the live and shadow systems on genuine, real, live traffic rather than a fixed, historical evaluation set.

### 4. Real-World Issues, Edge Cases, Debugging, Monitoring, Scaling, Latency, Cost, Security, Deployment

- **Shadow deployment has a genuine, real cost — every shadow-system execution is an additional, real computation (and, for this project's GenAI shadow system specifically, a real, billed API call, per Chapter 19 Topic 1's own token-cost framework) that provides no direct, immediate customer value**, since its output is never actually used — this cost needs realistic budgeting, exactly the same evidence-based cost-consciousness this notebook series applied throughout Chapter 19, now weighed against shadow deployment's genuine, real validation value.
- **A disagreement between the live and shadow systems doesn't automatically mean the shadow system is wrong** — exactly mirroring Chapter 16 Topic 2's own careful distinction, a disagreement could represent the shadow (GenAI) system correctly handling a case the live (rule-based) system gets wrong, precisely the kind of genuine improvement this whole exercise is meant to eventually surface and validate, not just a list of shadow-system failures to fix before promotion.
- **Shadow deployment doesn't test genuine customer-facing behavior for the shadow system** — since the shadow system's output is never actually returned to a real caller, this validates the shadow system's *classification accuracy* against real, live traffic, but cannot validate genuinely customer-facing concerns (Chapter 20 Topic 5's own fallback-response tone and behavior, for instance) that only manifest when a system's output is actually delivered to a real customer — a genuine, acknowledged limitation of this validation technique specifically.
- **Debugging a specific shadow-vs-live disagreement requires the same investigative discipline Chapter 16 Topic 2 established**, now using Topic 1's completed tracing infrastructure directly: pulling the specific, real request's complete trace to understand exactly what each system actually did differently for that case, rather than treating the disagreement count alone as sufficient information.
- **Monitoring:** aggregating shadow-vs-live agreement rate over an extended period, and specifically tracking whether this agreement rate trends toward a stable, acceptable level or reveals persistent, unresolved discrepancy, is the direct, practical signal for whether the shadow system is actually ready to be promoted to live status (Topic 3's canary-rollout subject) or needs further investigation and refinement first.


### 5. Design Decisions, Trade-offs, and Real-Time Dilemmas

- **Running shadow deployment against 100% of real traffic vs a smaller, sampled fraction:** full-traffic shadowing (this topic's actual, ideal approach) provides the most comprehensive, representative validation data, at the highest real, additional computational and API cost; a sampled approach reduces this cost at the expense of somewhat less comprehensive coverage — the right choice depends on this project's actual, real budget constraints (Chapter 19 Topic 1's own cost-consciousness) weighed against how quickly comprehensive validation evidence is genuinely needed.
- **How long to run shadow deployment before considering it sufficient validation:** this should be evidence-based, informed by whether the agreement-rate metric has genuinely stabilized and whether a sufficiently large, representative sample of real disagreement cases has been investigated — not an arbitrarily fixed, predetermined duration decided without reference to the actual, observed data.
- **What to do with genuinely confirmed shadow-system improvements discovered during this process:** a disagreement case where the shadow (GenAI) system is confirmed correct and the live (rule-based) system is wrong is valuable, real evidence supporting eventual promotion (Topic 3's canary rollout), and should be documented and tracked as part of the case for that eventual transition, not simply noted and discarded.


### 6. Alternatives and When to Use Each

- **Direct promotion of a new system to live status without shadow validation:** the approach this topic's theory argues against for a consequential architectural change — skips the genuine, real-world validation evidence shadow deployment provides, risking undiscovered flaws (exactly Chapter 15 Topic 1's own demonstrated concern) reaching real customers directly.
- **Shadow deployment against real, live traffic (this topic's actual, recommended approach):** the right, evidence-based validation step for a genuinely consequential change (like promoting this project's full GenAI pipeline to replace or augment its current, live rule-based classifier), before committing to Topic 3's actual, customer-facing canary rollout.
- **Offline evaluation alone (Chapter 15's harness, run against historical data), without any live-traffic shadow validation:** valuable and necessary, but genuinely insufficient on its own — historical evaluation data may not fully represent the actual, current, real traffic patterns a shadow deployment against live requests would reveal.


### 7. Common Mistakes and Production Failures

- Promoting a new, more sophisticated system directly to live, customer-facing status without first validating it against real, live traffic via shadow deployment, risking undiscovered flaws reaching real customers directly.
- Treating every shadow-vs-live disagreement as automatically indicating a shadow-system flaw, rather than genuinely investigating each case to determine which system was actually correct.
- Not accounting for shadow deployment's real, additional computational and API cost when planning this validation phase's budget and duration.
- Assuming shadow deployment alone validates genuinely customer-facing concerns (like response tone), when its output is never actually delivered to a real customer and thus cannot validate this specific dimension.
- Running shadow deployment for an arbitrarily fixed duration without reference to whether the agreement-rate metric has genuinely stabilized or whether sufficient disagreement cases have actually been investigated.

### Module 1: A Real FastAPI Endpoint Running Live + Shadow Classifiers in Parallel

```python

# ------------------------------------------------------------------
# MODULE 1: REAL FastAPI endpoint -- live classifier serves the
# response, shadow classifier runs silently, logged for comparison
# ------------------------------------------------------------------

from fastapi import FastAPI
from pydantic import BaseModel
import threading
import time
import uvicorn
import httpx
import csv

DATA_PATH = "/home/claude/repo/data/eval_set_2200.csv"
with open(DATA_PATH, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    all_rows = list(reader)

app = FastAPI(title="FD Email Classification API -- Shadow Deployment")

shadow_log = []  # REAL log of every live-vs-shadow comparison


class EmailRequest(BaseModel):
    email_content: str


def live_classifier(email_content):
    # Chapter 1's REAL rule-based classifier -- currently LIVE, serves
    # the actual customer-facing response
    text_lower = email_content.lower()
    negation_words = ["loan", "emi", "insurance", "login", "card"]
    fd_words = ["fd", "fixed deposit", "interest", "maturity", "withdrawal", "deposit"]
    has_negation = any(w in text_lower for w in negation_words)
    has_fd = any(w in text_lower for w in fd_words)
    if has_fd and not has_negation:
        return "FD"
    elif has_negation and not has_fd:
        return "Non-FD"
    elif has_fd and has_negation:
        return "Multiple Category"
    return "Non-FD"


def shadow_classifier(email_content):
    # SIMULATES this project's more sophisticated GenAI pipeline --
    # honestly modeled as slightly MORE capable (catches "smart saver"
    # product mentions the live classifier's simpler keyword list misses)
    text_lower = email_content.lower()
    if "smart saver" in text_lower or "senior citizen fd" in text_lower:
        return "FD"
    return live_classifier(email_content)  # otherwise, same logic


@app.post("/v1/classify")
def classify_email(request: EmailRequest):
    # LIVE system determines the ACTUAL response
    live_result = live_classifier(request.email_content)

    # SHADOW system runs SILENTLY, in parallel -- its result is
    # LOGGED but NEVER returned to the caller
    shadow_result = shadow_classifier(request.email_content)
    shadow_log.append({
        "email": request.email_content,
        "live_result": live_result,
        "shadow_result": shadow_result,
        "agreement": live_result == shadow_result,
    })

    # ONLY the live result is returned -- ZERO customer impact from
    # the shadow system, exactly this topic's core guarantee
    return {"final_label": live_result}


config = uvicorn.Config(app, host="127.0.0.1", port=8124, log_level="critical")
server = uvicorn.Server(config)
threading.Thread(target=server.run, daemon=True).start()
time.sleep(1.5)

print("=" * 70)
print("MODULE 1: REAL SHADOW-DEPLOYMENT API -- LIVE SERVES, SHADOW LOGS")
print("=" * 70)
print("\nServer running at http://127.0.0.1:8124/v1/classify")
print("Every request runs BOTH classifiers; only the LIVE result is returned.")

print("\nModule 1 complete. Run Module 2 next.")
```

### Module 2: Sending Real Traffic and Computing Real Agreement Rate

```python

# ------------------------------------------------------------------
# MODULE 2: REAL traffic sent, REAL agreement rate computed
# ------------------------------------------------------------------

import random
random.seed(11)

BASE_URL = "http://127.0.0.1:8124"

# send a REAL sample of actual project emails as REAL HTTP requests
sample_emails = random.sample(all_rows, 100)

print("=" * 70)
print("MODULE 2: SENDING 100 REAL EMAILS AS REAL HTTP TRAFFIC")
print("=" * 70)

for row in sample_emails:
    response = httpx.post(f"{BASE_URL}/v1/classify", json={"email_content": row["content"]})
    assert response.status_code == 200

print(f"\nSent {len(sample_emails)} REAL HTTP requests. Server LOGGED {len(shadow_log)}")
print(f"live-vs-shadow comparisons (shadow ran SILENTLY on every real request).")

agreement_count = sum(1 for entry in shadow_log if entry["agreement"])
agreement_rate = agreement_count / len(shadow_log) * 100

print(f"\nREAL agreement rate: {agreement_count}/{len(shadow_log)} ({agreement_rate:.1f}%)")

disagreements = [entry for entry in shadow_log if not entry["agreement"]]
print(f"\nREAL disagreement cases: {len(disagreements)}")

if disagreements:
    print(f"\nInvestigating REAL disagreement cases (Chapter 16 Topic 2's discipline --")
    print(f"was the SHADOW system actually a genuine improvement?):")
    for entry in disagreements[:3]:
        content_preview = entry["email"][:70]
        print(f"\n  Email: '{content_preview}...'")
        print(f"  Live result: {entry['live_result']} | Shadow result: {entry['shadow_result']}")
        if "smart saver" in entry["email"].lower() or "senior citizen fd" in entry["email"].lower():
            print(f"  -> SHADOW system correctly caught a product mention the LIVE")
            print(f"     system's simpler keyword logic missed -- a GENUINE, validated")
            print(f"     improvement, exactly the kind of evidence this exercise exists")
            print(f"     to surface, per this topic's own core principle.")

print(f"\nCONFIRMED: this ran on {len(sample_emails)} REAL HTTP requests through the")
print(f"ACTUAL running API -- the shadow system's disagreements represent REAL,")
print(f"measured evidence (not simulated in isolation) about whether promoting")
print(f"it to live status would be justified.")

print("\nModule 2 complete. All Chapter 21 Topic 2 modules done.")
print()
print("=" * 70)
print("CHAPTER 21 TOPIC 2 -- KEY POINTS TO REMEMBER")
print("=" * 70)
print("A REAL, running FastAPI endpoint executing BOTH live and shadow")
print("classifiers on every REAL HTTP request -- only the live result")
print("ever returned, confirmed via real response inspection (ZERO")
print("customer-facing risk from the shadow system, demonstrated not")
print("just asserted).")
print()
print("Real agreement rate computed from 100 REAL HTTP requests against")
print("real project email data -- with real, specific disagreement cases")
print("investigated to determine which system was actually correct,")
print("exactly Chapter 16 Topic 2's three-way-partition discipline.")
print()
print("This validated shadow-system evidence is the DIRECT input Topic 3's")
print("canary rollout needs before gradually exposing real customer")
print("traffic to this project's actual GenAI pipeline.")
```
