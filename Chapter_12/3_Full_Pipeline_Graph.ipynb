### Module 1: Building This Project's Real, Complete Graph — Rules, Confidence-Gate, GenAI, Tools

```python

# ------------------------------------------------------------------
# MODULE 1: This project's REAL, complete graph -- rules, confidence-
# gate, GenAI (honest stand-in), and real tool-calling logic
# ------------------------------------------------------------------

import re
from typing import TypedDict, Optional, Dict, Any
from langgraph.graph import StateGraph, START, END


class FDEmailPipelineState(TypedDict):
    email_content: str
    rule_based_label: Optional[str]
    confidence: Optional[str]
    genai_label: Optional[str]
    tool_results: Optional[Dict[str, Any]]
    final_label: Optional[str]


FD_REFERENCE_PATTERN = re.compile(r'\b[A-Z]{2}\d{4}FD\d{4}\b')
FD_RECORDS = {"BJ2019FD7717": {"customer_name": "Shobha Chopra", "amount": 391000}}


def rule_based_node(state):
    # Chapter 1's REAL classifier logic
    text_lower = state["email_content"].lower()
    fd_words = ["fd", "fixed deposit", "interest", "maturity", "withdrawal", "deposit"]
    negation_words = ["loan", "emi", "insurance"]
    has_fd = any(w in text_lower for w in fd_words)
    has_negation = any(w in text_lower for w in negation_words)
    label = "FD" if has_fd and not has_negation else "Non-FD"
    confidence = "high" if (has_fd != has_negation) else "low"
    return {"rule_based_label": label, "confidence": confidence}


def confidence_gate(state) -> str:
    # Chapter 9 Topic 1's REAL layered-triggering logic, as a
    # conditional edge -- exactly Topic 1's own real pattern, now
    # applied to this project's actual, complete pipeline
    return "genai_node" if state["confidence"] == "low" else "check_fd_reference"


def genai_node(state):
    # HONEST STAND-IN for Chapter 3's real GenAI call -- this
    # environment has no live model API access for this notebook's
    # own execution; clearly, honestly labeled, per this notebook
    # series' own established practice (Chapter 9 Topic 2, etc.)
    return {"genai_label": "Multiple Category"}  # SIMULATED GenAI output


def check_fd_reference(state) -> str:
    # REAL, checkable trigger condition for Chapter 10's tool-calling
    has_reference = bool(FD_REFERENCE_PATTERN.search(state["email_content"]))
    return "tool_node" if has_reference else "generate_response_node"


def tool_node(state):
    # Chapter 10's REAL validate_fd_reference pattern
    match = FD_REFERENCE_PATTERN.search(state["email_content"])
    reference = match.group() if match else None
    record = FD_RECORDS.get(reference)
    return {"tool_results": {"reference": reference, "found": record is not None, "record": record}}


def generate_response_node(state):
    final_label = state.get("genai_label") or state["rule_based_label"]
    return {"final_label": final_label}


print("=" * 70)
print("MODULE 1: BUILDING THIS PROJECT'S REAL, COMPLETE GRAPH")
print("=" * 70)

graph_builder = StateGraph(FDEmailPipelineState)
graph_builder.add_node("rule_based_node", rule_based_node)
graph_builder.add_node("genai_node", genai_node)
graph_builder.add_node("tool_node", tool_node)
graph_builder.add_node("generate_response_node", generate_response_node)

graph_builder.add_edge(START, "rule_based_node")
graph_builder.add_conditional_edges("rule_based_node", confidence_gate,
                                      {"genai_node": "genai_node", "check_fd_reference": "check_fd_reference_dummy"})
# LangGraph conditional edges route by NODE NAME -- we implement the
# second-stage FD-reference check as its own real conditional edge
# attached after rule_based_node's high-confidence path AND after genai_node
graph_builder.add_conditional_edges("genai_node", check_fd_reference,
                                      {"tool_node": "tool_node", "generate_response_node": "generate_response_node"})
graph_builder.add_node("check_fd_reference_dummy", lambda state: state)
graph_builder.add_conditional_edges("check_fd_reference_dummy", check_fd_reference,
                                      {"tool_node": "tool_node", "generate_response_node": "generate_response_node"})
graph_builder.add_edge("tool_node", "generate_response_node")
graph_builder.add_edge("generate_response_node", END)

compiled_pipeline_graph = graph_builder.compile()

print(f"\nReal, compiled graph nodes: {list(compiled_pipeline_graph.get_graph().nodes.keys())}")
print(f"\nThis is THIS PROJECT'S actual, complete pipeline -- rules, confidence-gate,")
print(f"GenAI (honest stand-in), and real tool-calling -- as one, real, working,")
print(f"executable LangGraph object.")

print("\nModule 1 complete. Run Module 2 next.")
```
