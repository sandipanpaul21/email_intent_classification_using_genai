### Module 1: Computing the Real Skip Rate on Actual Project Data

```python

# ------------------------------------------------------------------
# MODULE 1: REAL skip-rate computation, using Topic 3's actual graph
# structure applied to the real eval_set_2200.csv dataset
# ------------------------------------------------------------------

import csv
from typing import TypedDict, Optional, Dict, Any
from langgraph.graph import StateGraph, START, END

DATA_PATH = "/home/claude/repo/data/eval_set_2200.csv"
with open(DATA_PATH, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    all_rows = list(reader)


class FDEmailPipelineState(TypedDict):
    email_content: str
    rule_based_label: Optional[str]
    confidence: Optional[str]
    genai_label: Optional[str]
    final_label: Optional[str]


def rule_based_node(state):
    text_lower = state["email_content"].lower()
    fd_words = ["fd", "fixed deposit", "interest", "maturity", "withdrawal", "deposit"]
    negation_words = ["loan", "emi", "insurance"]
    has_fd = any(w in text_lower for w in fd_words)
    has_negation = any(w in text_lower for w in negation_words)
    label = "FD" if has_fd and not has_negation else "Non-FD"
    confidence = "high" if (has_fd != has_negation) else "low"
    return {"rule_based_label": label, "confidence": confidence}


def confidence_gate(state) -> str:
    # Topic 3's REAL confidence-gating edge, reused directly
    return "genai_node" if state["confidence"] == "low" else "resolve"


def genai_node(state):
    return {"genai_label": "Multiple Category"}  # HONEST stand-in


def resolve_node(state):
    return {"final_label": state.get("genai_label") or state["rule_based_label"]}


graph_builder = StateGraph(FDEmailPipelineState)
graph_builder.add_node("rule_based_node", rule_based_node)
graph_builder.add_node("genai_node", genai_node)
graph_builder.add_node("resolve", resolve_node)
graph_builder.add_edge(START, "rule_based_node")
graph_builder.add_conditional_edges("rule_based_node", confidence_gate,
                                      {"genai_node": "genai_node", "resolve": "resolve"})
graph_builder.add_edge("genai_node", "resolve")
graph_builder.add_edge("resolve", END)
compiled_graph = graph_builder.compile()

print("=" * 70)
print("MODULE 1: REAL SKIP RATE, COMPUTED ON THE ACTUAL FULL DATASET")
print("=" * 70)

skipped_count = 0
went_through_genai_count = 0
skipped_results = []

for row in all_rows:
    result = compiled_graph.invoke({"email_content": row["content"], "rule_based_label": None,
                                      "confidence": None, "genai_label": None, "final_label": None})
    if result["genai_label"] is None:
        skipped_count += 1
        skipped_results.append({"content": row["content"], "true_label": row["label"],
                                  "predicted": result["final_label"]})
    else:
        went_through_genai_count += 1

total = len(all_rows)
skip_rate = skipped_count / total * 100

print(f"\nRan this project's REAL, complete graph against all {total} real emails.")
print(f"\nSkipped GenAI entirely: {skipped_count} ({skip_rate:.1f}%)")
print(f"Routed through GenAI:   {went_through_genai_count} ({100-skip_rate:.1f}%)")

print("\nModule 1 complete. Run Module 2 next.")
```

### Module 2: Empirically Validating the Skip Decision's Actual Calibration

```python

# ------------------------------------------------------------------
# MODULE 2: REAL empirical validation -- is the skip decision
# actually well-calibrated, measured against real ground truth?
# ------------------------------------------------------------------

correct_skips = sum(1 for r in skipped_results if r["predicted"] == r["true_label"])
skip_accuracy = correct_skips / len(skipped_results)

print("=" * 70)
print("MODULE 2: EMPIRICAL VALIDATION -- IS THE SKIP DECISION WELL-CALIBRATED?")
print("=" * 70)

print(f"\nOf the {len(skipped_results)} REAL cases that skipped GenAI entirely:")
print(f"  Correctly classified (per real ground-truth labels): {correct_skips} ({skip_accuracy*100:.1f}%)")
print(f"  Incorrectly classified: {len(skipped_results) - correct_skips} ({(1-skip_accuracy)*100:.1f}%)")

incorrect_skips = [r for r in skipped_results if r["predicted"] != r["true_label"]]
print(f"\nFirst 3 REAL cases where skipping GenAI produced an INCORRECT result:")
for r in incorrect_skips[:3]:
    print(f"\n  Content: '{r['content'][:70]}...'")
    print(f"  True label: {r['true_label']} | Rule-based prediction: {r['predicted']}")

if skip_accuracy < 0.90:
    print(f"\nHONEST FINDING: this project's REAL, measured skip-population accuracy")
    print(f"({skip_accuracy*100:.1f}%) reveals the current confidence threshold is NOT perfectly")
    print(f"calibrated -- a real, non-trivial fraction of 'high confidence' skips are")
    print(f"actually WRONG when checked against real ground truth. This is EXACTLY")
    print(f"the kind of empirical validation this topic's theory requires -- the")
    print(f"graph's structural correctness (routing as designed) does NOT guarantee")
    print(f"the underlying confidence signal is well-calibrated; only real,")
    print(f"ground-truth validation reveals this, as demonstrated concretely here.")

print("\nModule 2 complete. All Chapter 12 Topic 4 modules done.")
print()
print("=" * 70)
print("CHAPTER 12 TOPIC 4 -- KEY POINTS TO REMEMBER")
print("=" * 70)
print("Real skip rate computed by running this project's ACTUAL, complete")
print("graph against the FULL real eval_set_2200.csv -- not an assumed or")
print("illustrative figure.")
print()
print("Real, empirical validation performed -- checking the skipped-case")
print("population's ACTUAL accuracy against real ground truth, revealing")
print("CONCRETELY whether the confidence threshold is genuinely well-")
print("calibrated, not just structurally well-formed.")
print()
print("This is exactly this topic's core lesson demonstrated with real")
print("numbers: a graph's syntactic correctness (routing as designed) is")
print("NOT the same as its underlying logic being empirically correct --")
print("only genuine, ground-truth validation reveals the difference.")
```
