```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize


data = pd.read_csv("../data/fd_dataset_messy.csv")
print(data.head(5))
```

```python
# inspect tokenized content and add a token length feature
data['token_count'] = data['content_token'].apply(len)

print(data[['content', 'content_token', 'token_count']].head(5))

sample_idx = 0
print("\nSample text:")
print(data.loc[sample_idx, 'content'])
print("\nSample tokens:")
print(data.loc[sample_idx, 'content_token'])
```

```python
content_token = []

for text in data['content']:
    tokens = word_tokenize(str(text))
    content_token.append(tokens)

data['content_token'] = content_token
print("First Email Content:", data['content'][0])
print("First Email Content Tokens:", content_token[0])
```
