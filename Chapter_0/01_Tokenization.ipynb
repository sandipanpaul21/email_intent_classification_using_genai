```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize


data = pd.read_csv("../data/fd_dataset_messy.csv")
print(data.head(5))
```

```python
# lowercase right after reading
data['content'] = data['content'].str.lower()
data['subject'] = data['subject'].str.lower()

data.head(5)
```

##### Word Tokenizer

```python
content_token = []

for text in data['content']:
    tokens = word_tokenize(str(text))
    content_token.append(tokens)

data['content_token'] = content_token
print("First Email Content:", data['content'][0])
print("\n")
print("First Email Content Tokens:", content_token[0])
```

```python
print(data.head(5))
```

##### Character Tokenizer

```python
ch_token = []

for text in data['content']:
    tokens  =[]
    for ch in str(text):
        if ch.isalpha():    # ch.isalpha() is True (skipping spaces, punctuation, digits, etc.).
            tokens.append(ch)
    ch_token.append(tokens)

print("First Email Content:", data['content'][0])
print("First Email Alpha-Only Char Tokens:", ch_token[0])
print("\n")
data['content_char_token'] = ch_token
```

##### BPE Tokenizer used in GPT

Important Points
- BPE(unk_token="[UNK]") 
    - This creates the actual algorithm (the BPE model) that will learn and apply merge rules. 
    - The unk_token="[UNK]" part tells it: "if you ever see a token during tokenization that wasn't learned in training, replace it with this placeholder instead of crashing." 
    - UNK = unknown.

- Whitespace()
    - Before BPE can merge characters into subwords, it needs initial word boundaries — otherwise it might try merging characters across two different words. 
    - Whitespace() splits text on spaces and separates punctuation into its own piece. 

- train_from_iterator
    - This is where the real BPE algorithm runs, following exactly the flow from your cheatsheet:
    1. Starts by breaking every pre-tokenized chunk into individual characters ("branch" → b,r,a,n,c,h).
    2. Scans your whole data['content'] column and counts which adjacent character pairs occur most often.
    3. Repeatedly merges the most frequent pair into a single new subword, then recounts, then merges again.
    4. Stops once the vocabulary reaches vocab_size=1000 total subwords.
    5. special_tokens=["[UNK]"] reserves a slot in that vocabulary specifically for [UNK] — without this, the earlier error you saw (Unk token not found) happens, since [UNK] wouldn't exist in the learned vocabulary at all.

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

bpe_tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
bpe_tokenizer.pre_tokenizer = Whitespace()
bpe_tokenizer.train_from_iterator(data['content'], BpeTrainer(vocab_size=1000, special_tokens=["[UNK]"]))

bpe_token = []
for text in data['content']:
    tokens = bpe_tokenizer.encode(str(text)).tokens
    bpe_token.append(tokens)

data['bpe_token'] = bpe_token
print("First Email Content:", data['content'][0])
print("\n")
print("First Email BPE Tokens:", bpe_token[0])
```

##### Word Piece Tokenizer used in BERT

```python
from tokenizers import Tokenizer
from tokenizers.models import WordPiece
from tokenizers.trainers import WordPieceTrainer
from tokenizers.pre_tokenizers import Whitespace

wp_tokenizer = Tokenizer(WordPiece(unk_token="[UNK]"))
wp_tokenizer.pre_tokenizer = Whitespace()
wp_tokenizer.train_from_iterator(data['content'], WordPieceTrainer(vocab_size=1000, special_tokens=["[UNK]"]))

wordpiece_token = []
for text in data['content']:
    tokens = wp_tokenizer.encode(str(text)).tokens
    wordpiece_token.append(tokens)

data['wordpiece_token'] = wordpiece_token
print("First Email Content:", data['content'][0])
print("\n")
print("First Email WordPiece Tokens:", wordpiece_token[0])
```

##### Sentence Piece Tokenizer

Important Points

- model_prefix="sp_model" 
    - This is the important part. Unlike the HuggingFace tokenizers library (which keeps the trained tokenizer as a Python object in memory, e.g. bpe_tokenizer), SentencePiece writes the trained model straight to disk as two files: sp_model.model and sp_model.vocab.

- sp = spm.SentencePieceProcessor() and sp.load("sp_model.model")
    - This is the step that's easy to miss the purpose of: spm.SentencePieceTrainer.train(...) only creates the files — it doesn't hand you back a ready-to-use tokenizer object. 
    - So you create a fresh, empty SentencePieceProcessor(), then .load("sp_model.model") reads the trained model file back off disk into that object. 

```python
import sentencepiece as spm

spm.SentencePieceTrainer.train(
    sentence_iterator=iter(data['content']),
    model_prefix="sp_model",
    vocab_size=1000,
    model_type="bpe"
)

sp = spm.SentencePieceProcessor()
sp.load("sp_model.model")

sentencepiece_token = []
for text in data['content']:
    tokens = sp.encode(str(text), out_type=str)
    sentencepiece_token.append(tokens)

data['sentencepiece_token'] = sentencepiece_token
print("First Email Content:", data['content'][0])
print("\n")
print("First Email SentencePiece Tokens:", sentencepiece_token[0])
```
