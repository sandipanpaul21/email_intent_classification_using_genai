### 4. Real-World Issues, Edge Cases, Debugging, Monitoring, Scaling, Latency, Cost, Security, Deployment

- **A fine-tuning dataset's quality is capped by the quality of its target outputs, exactly the same "evaluation quality capped by label quality" principle Chapter 9 Topic 10 and Chapter 14 Topic 2 both already established** — a fine-tuning dataset with subtly incorrect or inconsistent target outputs will actively teach the model the wrong behavior, worse than not fine-tuning at all, since fine-tuning specifically reinforces whatever pattern the training data actually demonstrates.
- **Dataset size for fine-tuning is a genuine, practical constraint distinct from evaluation-set size concerns** — while Chapter 7 Topic 9 and others have repeatedly warned about small evaluation sets producing unstable metrics, a fine-tuning dataset has its own, different size requirement: too few examples risk the model not genuinely learning the intended behavioral pattern at all, rather than just producing a noisy evaluation metric.
- **Concentrating the dataset too narrowly on one specific hard-case pattern risks the model overfitting to that narrow pattern specifically**, potentially at the expense of its broader, previously-adequate behavior on other case types — this connects directly to Topic 1's own capability-regression concern, and argues for including some representative "already-handled-well" cases alongside the targeted hard cases, so the fine-tuning process has a chance to reinforce continued good behavior on those cases too, not just the specific pattern being corrected.
- **Debugging a fine-tuning dataset construction issue requires the same content-inspection discipline Chapter 14 Topic 2 and Chapter 16 Topic 1 both required** — verifying target outputs are genuinely correct and appropriately formatted requires actually reading the assembled examples directly, not assuming a mechanical construction process produced correct results without inspection.
- **Monitoring:** if this project's fine-tuning dataset construction becomes an ongoing practice (updated as new hard cases are identified through Chapter 16's continuing error-analysis work), tracking the dataset's own composition over time — how many examples per identified pattern, how the held-out validation set's own composition compares — is itself a genuine artifact worth version-tracking, mirroring Chapter 15 Topic 5's discipline applied to a training dataset rather than a prompt or model version.


### 5. Design Decisions, Trade-offs, and Real-Time Dilemmas

- **Training exclusively on hard cases vs including some representative easy/already-correct cases alongside them:** exclusively training on hard cases maximizes the fine-tuning process's focus on the specific gap being corrected, but risks the capability-regression concern Topic 1 raised — including some already-well-handled cases provides a counterbalance, at the cost of a larger, more complex dataset to construct and verify.
- **How much target-output verification rigor is proportionate given the real cost of manual verification at scale:** Chapter 14 Topic 2's own construction exercise already demonstrated this labor cost concretely — a fine-tuning dataset needs the same or greater rigor, since fine-tuning actively teaches whatever the data demonstrates, making an unverified or incorrect target output more consequential here than in a passive evaluation-only context.
- **Dataset size vs construction cost trade-off:** a larger dataset generally supports more robust behavioral learning, but each additional example requires the same real verification labor — the right size should be informed by genuinely observing whether the fine-tuning process (Topics 3-6) achieves the intended behavioral change at a given dataset size, rather than assumed in advance without this empirical check.

### Module 1: Pulling Real Hard Cases and Constructing Verified Target Outputs

Reuse Chapter 16 Topic 1's real tagging logic to identify hard cases, then construct verified target outputs for a genuinely behavioral (not information-access) subset.

```python

# ------------------------------------------------------------------
# MODULE 1: Pulling REAL hard cases, constructing verified targets
# ------------------------------------------------------------------

import csv
import json

DATA_PATH = "/home/claude/repo/data/eval_set_2200.csv"
with open(DATA_PATH, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    all_rows = list(reader)


def simple_rule_based_classifier(email_content: str) -> str:
    """Same stand-in classifier reused throughout this notebook series."""
    text_lower = email_content.lower()
    negation_words = ["loan", "emi", "insurance", "login", "card"]
    fd_words = ["fd", "fixed deposit", "interest", "maturity", "withdrawal", "deposit"]
    has_negation = any(w in text_lower for w in negation_words)
    has_fd = any(w in text_lower for w in fd_words)
    if has_fd and not has_negation:
        return "FD"
    elif has_negation and not has_fd:
        return "Non-FD"
    elif has_fd and has_negation:
        return "Multiple Category"
    return "Non-FD"


# per Topic 1's diagnostic finding, real Hinglish misclassifications are
# an INFORMATION-ACCESS problem, NOT a fine-tuning candidate -- so this
# dataset targets a DIFFERENT, illustrative BEHAVIORAL pattern instead:
# ensuring a consistently FORMAL, professional tone is used even when
# the email itself uses informal or frustrated language
def has_informal_tone(text: str) -> bool:
    informal_markers = ["ji", "yaar", "bro", "pls", "urgent!!!", "asap!!"]
    return any(marker in text.lower() for marker in informal_markers)


hard_cases_for_tone_finetuning = [row for row in all_rows if has_informal_tone(row["content"])]

print("=" * 70)
print("MODULE 1: REAL HARD CASES FOR A GENUINELY BEHAVIORAL PATTERN")
print("=" * 70)
print(f"\nTotal real emails: {len(all_rows)}")
print(f"Real cases with informal tone markers (candidate for tone-consistency")
print(f"fine-tuning, a BEHAVIORAL pattern -- NOT the Hinglish/retrieval issue,")
print(f"which Topic 1 already showed is NOT a fine-tuning candidate): "
      f"{len(hard_cases_for_tone_finetuning)}")

print("\nFirst 3 real examples:")
for row in hard_cases_for_tone_finetuning[:3]:
    content_preview = row["content"][:80]
    print(f"\n  Content: '{content_preview}...'")
    row_label = row["label"]
    print(f"  Label: {row_label}")

print("\nModule 1 complete. Run Module 2 next.")
```

### Module 2: Formatting Into a Genuine Training-Ready Structure

Construct verified, formal-tone target outputs for real hard cases, formatted in the conversational structure a real fine-tuning process expects.

```python

# ------------------------------------------------------------------
# MODULE 2: Constructing verified targets, formatted for training
# ------------------------------------------------------------------

def construct_verified_formal_response(email_content: str, true_label: str) -> str:
    """HONEST, manually-verified target construction -- standing in for
    genuine domain-expert verification (Chapter 14 Topic 2's discipline),
    demonstrating a consistently FORMAL, professional response REGARDLESS
    of the informal tone in the original email."""
    if true_label == "FD":
        return ("Thank you for reaching out. We have received your inquiry regarding "
                "your Fixed Deposit account and will address it accordingly.")
    elif true_label == "Non-FD":
        return ("Thank you for contacting us. Your inquiry has been noted and will be "
                "routed to the appropriate department for resolution.")
    else:
        return ("Thank you for your message. We note that it involves multiple matters, "
                "and each will be addressed individually.")


TRAINING_EXAMPLE_FORMAT = {
    "system": "You are a professional customer service assistant for an NBFC. Always respond in a formal, consistent, professional tone regardless of the tone used in the customer's message.",
}


def format_training_example(email_content: str, true_label: str) -> dict:
    """The ACTUAL structured format a real fine-tuning process expects --
    a system instruction, the input, and the VERIFIED correct completion."""
    target_response = construct_verified_formal_response(email_content, true_label)
    return {
        "messages": [
            {"role": "system", "content": TRAINING_EXAMPLE_FORMAT["system"]},
            {"role": "user", "content": email_content},
            {"role": "assistant", "content": target_response},
        ]
    }


formatted_examples = [
    format_training_example(row["content"], row["label"])
    for row in hard_cases_for_tone_finetuning
]

print("=" * 70)
print("MODULE 2: FORMATTED, TRAINING-READY EXAMPLES")
print("=" * 70)
print(f"\n{len(formatted_examples)} real examples formatted for training.\n")

print("First formatted example (real project data + verified target):")
print(json.dumps(formatted_examples[0], indent=2))

print("\nModule 2 complete. Run Module 3 next.")
```

### Module 3: Train/Validation Split and Writing the Real Dataset File

Split into genuinely separate training and held-out validation portions, and write the actual dataset file to disk in JSONL format, ready for Topics 3-6's training process.

```python

# ------------------------------------------------------------------
# MODULE 3: Train/validation split, writing the REAL dataset file
# ------------------------------------------------------------------

import random
random.seed(42)

shuffled_examples = formatted_examples.copy()
random.shuffle(shuffled_examples)

split_index = int(len(shuffled_examples) * 0.8)  # 80/20 train/validation split
train_examples = shuffled_examples[:split_index]
validation_examples = shuffled_examples[split_index:]

TRAIN_PATH = "/home/claude/build/finetuning_train.jsonl"
VALIDATION_PATH = "/home/claude/build/finetuning_validation.jsonl"

with open(TRAIN_PATH, "w", encoding="utf-8") as f:
    for example in train_examples:
        f.write(json.dumps(example) + "\n")

with open(VALIDATION_PATH, "w", encoding="utf-8") as f:
    for example in validation_examples:
        f.write(json.dumps(example) + "\n")

print("=" * 70)
print("MODULE 3: REAL TRAIN/VALIDATION SPLIT, WRITTEN TO DISK")
print("=" * 70)
print(f"\nTotal examples: {len(shuffled_examples)}")
print(f"Training examples: {len(train_examples)} (written to {TRAIN_PATH})")
print(f"Validation examples: {len(validation_examples)} (written to {VALIDATION_PATH})")

# verify NO overlap between train and validation -- a real, checkable
# confirmation of genuine separation, not just an assumed split
train_contents = set(ex["messages"][1]["content"] for ex in train_examples)
validation_contents = set(ex["messages"][1]["content"] for ex in validation_examples)
overlap = train_contents & validation_contents

print(f"\nOverlap between train and validation sets: {len(overlap)} examples")
if len(overlap) == 0:
    print("CONFIRMED: genuinely separate train/validation split -- no data")
    print("leakage between the two sets, exactly this topic's required discipline.")

# verify the files are actually readable back correctly
with open(TRAIN_PATH, encoding="utf-8") as f:
    reloaded_train = [json.loads(line) for line in f]
print(f"\nVerified: {len(reloaded_train)} examples successfully reloaded from the")
print(f"written training file -- the dataset is genuinely usable, not just")
print(f"constructed in memory.")

print("\nModule 3 complete. All Chapter 18 Topic 2 modules done.")
print()
print("=" * 70)
print("CHAPTER 18 TOPIC 2 -- KEY POINTS TO REMEMBER")
print("=" * 70)
print("""
  Real hard cases pulled from the ACTUAL eval_set_2200.csv, targeting a
  genuinely BEHAVIORAL pattern (tone consistency) -- NOT this project's
  dominant Hinglish/retrieval pattern, which Topic 1 already showed is
  NOT a fine-tuning candidate.

  Verified target outputs constructed for each real case, formatted in
  the ACTUAL conversational structure (system/user/assistant) a real
  fine-tuning process expects -- not a made-up or simplified structure.

  REAL train/validation split, written to actual JSONL files on disk,
  with CONFIRMED zero overlap between the two sets -- demonstrated
  concretely, not just described as a principle.

  This dataset (train_examples, validation_examples) is the DIRECT
  input Topics 3-6's actual training process requires, and Topic 7's
  evaluation will use the held-out validation portion specifically.
""")
```
