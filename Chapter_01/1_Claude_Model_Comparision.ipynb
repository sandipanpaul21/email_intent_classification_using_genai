NOTE ON MODEL IDs
Opus 4.8 and Sonnet 4.6 use a **dateless, pinned** model ID (not an evergreen pointer) — so unlike older generations, the API ID and the alias are literally the same string. Haiku 4.5 still follows the older **dated-snapshot** convention, so its API ID and alias differ.

##### **CLAUDE HAIKU 4.5 — FEATURE BREAKDOWN (WHY EACH SPEC MATTERS)**

- **Pricing:**
  - What it means: Billed separately for input tokens (what you send) and output tokens (what it generates) — output costs more because generation is more compute-intensive than reading.
  - Advantage: Cheapest tier today — classifying all 2,500 emails in your dataset costs roughly $0.36 end to end. Good fit for high-volume, low-margin-per-call work.
  - Disadvantage: Cheap because the model is smaller — misclassifies nuanced cases (like "Multiple Category" emails) more often than Sonnet/Opus.
  - What to watch for: Cost = price x volume. An unoptimized, repeated system prompt without caching can make the "cheap" model expensive at scale anyway.

- **Extended thinking:** Yes
  - What it means: Optional budgeted reasoning mode — Claude drafts internal reasoning steps before the final answer; you cap how many tokens it can spend doing that.
  - So turning Extended Thinking ON doesn't add a separate "thinking fee" — it just means the model generates more output tokens (the reasoning steps), and all output tokens bill at the output rate, which is already the expensive side of the pricing table (e.g., Sonnet's $15/MTok output vs $3/MTok input — a 5x multiplier before thinking even enters the picture)
  - Advantage: Better accuracy on ambiguous inputs without upgrading models — e.g., correctly splitting a mixed loan-complaint-plus-FD-query email.
  - Disadvantage: Thinking tokens bill at output rates whether or not they're displayed — turning this on quietly raises cost and latency.
  - What to watch for: Off by default — turn it on selectively, only for the input patterns you've measured Haiku struggling with, not globally.

- **Adaptive thinking:** No
  - What it means: Haiku lacks the "model decides its own reasoning depth" mode available only on Sonnet 4.6 and Opus 4.8 — only the manual toggle above exists here.
  - Advantage: Predictable cost and behavior — you decide exactly when reasoning kicks in, not the model.
  - Disadvantage: No automatic effort-scaling — requires hand-tuning per request type, extra engineering work.
  - What to watch for: Fine for "manual control" workflows; not the right model if you want "set it and forget it."

- **Comparative latency:** Fastest
  - What it means: Returns a full response quickest relative to other current Claude models for a similar-sized request.
  - Advantage: Best fit for real-time, user-facing moments — e.g., a live suggestion box as a support agent opens an email.
  - Disadvantage: Speed and depth trade off industry-wide — the fastest model is also most likely to miss subtlety.
  - What to watch for: "Fastest" is relative and payload-dependent — benchmark with your actual prompt size, not the marketing label.

- **Context window:** 200k tokens
  - What it means: Max input size per request — roughly 150k words.
  - tokens ≠ words. So, 200k tokens ~ 150k words
  - Advantage: Your full dataset (~133k tokens) fits in a single call with room to spare.
  - Disadvantage: Smallest window in the current lineup (Sonnet/Opus get 1M) — hits the ceiling first if email volume grows 5–10x.
  - What to watch for: Context window isn't memory — it resets every call. Continuity across calls means resending context, which adds straight back onto your token bill.

- **Max output:** 64k tokens
  - What it means: Hard ceiling on how much a single response can generate.
  - Advantage: Comfortably covers your batch job (~45k tokens for label + short reason across all 2,500 emails) in one call.
  - Disadvantage: Longer per-email explanations instead of short tags could exceed 64k and get cut off or need splitting across calls.
  - What to watch for: Do the output-token math against your actual batch size before shipping — this is exactly the kind of limit that causes a job to silently truncate in production.
