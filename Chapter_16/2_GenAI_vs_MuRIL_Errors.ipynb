## Part A: Theory

### 1. Concept, Intuition, and Why This Exists

- This project's entire GenAI-augmented pipeline exists specifically to beat a stated baseline: the MuRIL model achieving 0.97 FD Recall, referenced repeatedly throughout this notebook series as the benchmark this project's more sophisticated architecture is meant to improve upon. Topic 1 built a real, tagged taxonomy of this project's own current failure patterns — this topic asks the natural, comparative question that taxonomy alone can't answer: are the GenAI-augmented system's errors the *same kind* of errors MuRIL made, or has moving to a more sophisticated architecture actually changed *what kind* of mistakes get made, not just how many?
- This distinction matters enormously for deciding what to do next: if the GenAI system fails on the exact same cases and for the exact same underlying reasons as MuRIL, the more sophisticated architecture hasn't actually solved the hard cases — it may have just shifted the same errors elsewhere or reduced their frequency without addressing their root cause. If the GenAI system fails on genuinely *different* cases, for genuinely *different* reasons, this reveals something valuable about what each architecture is actually good and bad at — information no aggregate accuracy comparison alone could reveal.
- The project's own stated candidate MuRIL-specific error patterns — short emails, sarcasm, new products — are exactly the kind of hypothesis Topic 1's own tagging discipline should be applied to verify directly, not assumed true by reputation; this topic's contribution is the actual, side-by-side comparison of Topic 1's real, tagged GenAI failure taxonomy against a similarly-constructed taxonomy of MuRIL's own real failures on the same underlying data.


### 2. Internal Working — Step by Step

**Performing a genuine, evidence-based error-pattern comparison, step by step:**

1. **Apply Topic 1's exact same tagging methodology to MuRIL's own misclassifications** on the same underlying evaluation data — using the identical failure-pattern taxonomy (or a deliberately extended version, if MuRIL reveals patterns the GenAI taxonomy didn't need) ensures the comparison is genuinely apples-to-apples, rather than comparing two differently-constructed, incompatible analyses.
2. **Compute the overlap between the two systems' failure sets at the level of individual examples**, not just pattern frequencies — specifically, how many emails does *both* systems get wrong, versus how many does only MuRIL get wrong, versus how many does only the GenAI system get wrong? This three-way partition is the concrete, example-level evidence for whether the two systems' failures are the same cases or genuinely different ones.
3. **Compare failure-pattern *frequency profiles* between the two systems**, using Topic 1's aggregated tag-frequency methodology applied separately to each system's failures — if "short email lacking context" dominates MuRIL's errors but is a minor factor in the GenAI system's errors, this is direct, quantified evidence that the more sophisticated architecture has genuinely changed what kind of case is hard, not just how often failures occur.
4. **Investigate cases in the "GenAI-only" failure set specifically** — cases MuRIL handled correctly but the GenAI system gets wrong — since this is the specific, concerning category that represents a genuine regression introduced by the architectural change, deserving focused attention distinct from cases both systems fail on (likely reflecting genuinely hard, shared task difficulty) or cases only MuRIL failed on (representing the GenAI system's actual, realized improvement).

### 4. Real-World Issues, Edge Cases, Debugging, Monitoring, Scaling, Latency, Cost, Security, Deployment

- **A "GenAI-only" failure — a case MuRIL handled correctly but the GenAI system gets wrong — is a genuine, concerning regression that deserves focused, prioritized investigation**, distinct from a shared failure both systems exhibit (likely reflecting genuinely hard task difficulty neither architecture has solved) — conflating these two categories risks either under-investigating a real, architecture-introduced regression or over-investigating a case that may simply reflect inherent task difficulty no architecture can fully eliminate.
- **Comparing aggregate accuracy alone between the two systems (this project's baseline 0.97 Recall vs the GenAI system's own measured accuracy) hides exactly the information this topic's comparison exists to reveal** — two systems could have identical aggregate accuracy while failing on completely different cases for completely different reasons, information only this kind of example-level, pattern-level comparison surfaces.

- **Debugging a case where both systems fail requires different investigation than debugging a GenAI-only failure** — a shared failure might point toward a genuinely hard, perhaps irreducible task-difficulty case (worth documenting as an accepted baseline error rate rather than endlessly pursued), while a GenAI-only failure points toward something the new architecture specifically introduced or failed to handle that the simpler, prior approach handled adequately.

### 6. Alternatives and When to Use Each

- **Aggregate accuracy comparison alone (MuRIL's 0.97 Recall vs the GenAI system's own measured accuracy):** a reasonable, simple headline comparison, but insufficient on its own for understanding whether the underlying nature of failures has actually changed — exactly this topic's core argument.
- **Full, example-level and pattern-level comparative error analysis (this topic's actual approach):** the right choice for genuinely understanding whether a new architecture has solved the same problems differently, solved genuinely different problems, or simply shifted where the same fundamental difficulties surface.
- **Comparing only pattern frequencies without the example-level three-way partition:** a useful, lighter-weight intermediate step, but misses the specific, most actionable finding (which exact cases represent genuine GenAI-introduced regressions) that only example-level comparison reveals.


### 7. Common Mistakes and Production Failures

- Relying solely on aggregate accuracy comparison between MuRIL and the GenAI system, missing whether the underlying nature of failures has genuinely changed or merely shifted in frequency.
- Not specifically isolating and investigating "GenAI-only" failures (cases MuRIL handled correctly but the new system doesn't), missing genuine, architecture-introduced regressions hidden within an otherwise-improved aggregate accuracy figure.
- Treating a shared failure (both systems wrong) with the same urgency as a GenAI-only regression, rather than recognizing that shared failures may reflect genuinely hard, potentially irreducible task difficulty deserving a different kind of response.
- Presenting a simulated or illustrative MuRIL-comparison exercise's specific findings as if they were validated conclusions about the project's actual, real MuRIL baseline, rather than honestly scoping such an exercise as a methodology demonstration.

### Module 1: A Simulated MuRIL-Style Classifier With Its Own Real, Distinct Failure Tendencies

Build a simulated MuRIL-like classifier honestly labeled as a stand-in, designed to plausibly fail on this project's stated candidate patterns (short emails, sarcasm) -- run against the actual full dataset.

```python

# ------------------------------------------------------------------
# MODULE 1: Simulated MuRIL-style classifier (honest stand-in) +
# the GenAI stand-in classifier from Topic 1, BOTH run on REAL data
# ------------------------------------------------------------------

import csv

DATA_PATH = "/home/claude/repo/data/eval_set_2200.csv"
with open(DATA_PATH, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    all_rows = list(reader)


def genai_classifier(email_content: str) -> str:
    """SAME GenAI stand-in classifier from Topic 1, unchanged."""
    text_lower = email_content.lower()
    negation_words = ["loan", "emi", "insurance", "login", "card"]
    fd_words = ["fd", "fixed deposit", "interest", "maturity", "withdrawal", "deposit"]
    has_negation = any(w in text_lower for w in negation_words)
    has_fd = any(w in text_lower for w in fd_words)
    if has_fd and not has_negation:
        return "FD"
    elif has_negation and not has_fd:
        return "Non-FD"
    elif has_fd and has_negation:
        return "Multiple Category"
    return "Non-FD"


def simulated_muril_classifier(email_content: str) -> str:
    """A SIMULATED, honestly-labeled stand-in for the real MuRIL model
    (not accessible in this environment) -- deliberately designed to
    plausibly exhibit this project's OWN STATED candidate MuRIL failure
    patterns: struggling specifically on SHORT emails (insufficient
    context for a single-pass classifier with no reasoning step) and
    on emails using INDIRECT/SARCASTIC phrasing that a simpler,
    non-reasoning classifier would read too literally."""
    word_count = len(email_content.split())
    text_lower = email_content.lower()

    # SIMULATED weakness 1: short emails -- MuRIL-style classifier
    # defaults to a WEAKER guess when there isn't much text to work with
    if word_count <= 8:
        return "Non-FD"  # a plausible, but frequently WRONG, default guess

    # SIMULATED weakness 2: sarcastic/indirect phrasing read too literally
    sarcasm_markers = ["great job", "wonderful service", "thanks a lot", "really appreciate"]
    if any(marker in text_lower for marker in sarcasm_markers):
        return "Non-FD"  # reads the surface-positive phrasing as a non-issue

    # otherwise, a reasonably competent keyword-based guess (MuRIL is a
    # real, fairly strong baseline -- NOT a strawman)
    fd_words = ["fd", "fixed deposit", "interest", "maturity", "withdrawal", "deposit"]
    negation_words = ["loan", "emi", "insurance", "login", "card"]
    has_fd = any(w in text_lower for w in fd_words)
    has_negation = any(w in text_lower for w in negation_words)
    if has_fd and not has_negation:
        return "FD"
    elif has_negation:
        return "Non-FD"
    return "Non-FD"


genai_results = []
muril_results = []
for row in all_rows:
    genai_pred = genai_classifier(row["content"])
    muril_pred = simulated_muril_classifier(row["content"])
    genai_results.append({"content": row["content"], "true_label": row["label"],
                            "predicted": genai_pred, "correct": genai_pred == row["label"]})
    muril_results.append({"content": row["content"], "true_label": row["label"],
                            "predicted": muril_pred, "correct": muril_pred == row["label"]})

genai_accuracy = sum(r["correct"] for r in genai_results) / len(genai_results)
muril_accuracy = sum(r["correct"] for r in muril_results) / len(muril_results)

print("=" * 70)
print("MODULE 1: BOTH CLASSIFIERS RUN ON THE REAL, FULL DATASET")
print("=" * 70)
print(f"\nGenAI stand-in accuracy:  {genai_accuracy:.3f}")
print(f"MuRIL stand-in accuracy:  {muril_accuracy:.3f}")
print(f"\n(These are ILLUSTRATIVE stand-ins, not the project's real MuRIL")
print(f"0.97-Recall baseline -- this exercise demonstrates the COMPARISON")
print(f"METHODOLOGY using real data, honestly labeled as simulated models.)")

print("\nModule 1 complete. Run Module 2 next.")
```

### Module 2: The Three-Way Partition — Both-Wrong, MuRIL-Only, GenAI-Only

Compute the real, example-level three-way partition across all 2,200 real emails, revealing whether the two systems fail on the same or genuinely different cases.

```python

# ------------------------------------------------------------------
# MODULE 2: Real, example-level three-way partition
# ------------------------------------------------------------------

both_wrong = []
muril_only_wrong = []
genai_only_wrong = []

for genai_r, muril_r in zip(genai_results, muril_results):
    genai_wrong = not genai_r["correct"]
    muril_wrong = not muril_r["correct"]

    if genai_wrong and muril_wrong:
        both_wrong.append(genai_r)
    elif muril_wrong and not genai_wrong:
        muril_only_wrong.append(genai_r)
    elif genai_wrong and not muril_wrong:
        genai_only_wrong.append(genai_r)

print("=" * 70)
print("MODULE 2: THREE-WAY PARTITION -- REAL, EXAMPLE-LEVEL COMPARISON")
print("=" * 70)

total = len(all_rows)
print(f"\nTotal emails: {total}")
print(f"Both systems WRONG (likely shared, genuine task difficulty): {len(both_wrong)} ({len(both_wrong)/total*100:.1f}%)")
print(f"MuRIL-only WRONG (GenAI system's realized improvement):      {len(muril_only_wrong)} ({len(muril_only_wrong)/total*100:.1f}%)")
print(f"GenAI-only WRONG (genuine, concerning NEW regression):       {len(genai_only_wrong)} ({len(genai_only_wrong)/total*100:.1f}%)")

print(f"\nNet effect: GenAI system FIXED {len(muril_only_wrong)} cases MuRIL got wrong,")
print(f"but INTRODUCED {len(genai_only_wrong)} new regressions MuRIL handled correctly.")

if len(genai_only_wrong) > 0:
    print(f"\nFirst 3 real GenAI-ONLY regression examples (MuRIL got these RIGHT):")
    for ex in genai_only_wrong[:3]:
        content_preview = ex["content"][:70]
        ex_true = ex["true_label"]
        ex_pred = ex["predicted"]
        print(f"\n  True: {ex_true} | GenAI predicted: {ex_pred}")
        print(f"  Content: '{content_preview}...'")

print("\nModule 2 complete. Run Module 3 next.")
```

### Module 3: Comparing Failure-Pattern Frequency Profiles Between the Two Systems

Apply Topic 1's exact tagging methodology to BOTH systems' failures, comparing which patterns dominate each -- revealing whether the architectures genuinely differ in what kind of case is hard.

```python

# ------------------------------------------------------------------
# MODULE 3: Comparing failure-pattern PROFILES between both systems
# ------------------------------------------------------------------

from collections import Counter

def tag_failure_patterns(email_content: str) -> list:
    """EXACT SAME tagging function from Topic 1, applied here to BOTH
    systems' failures for a genuinely apples-to-apples comparison."""
    tags = []
    text_lower = email_content.lower()
    word_count = len(email_content.split())

    if word_count <= 8:
        tags.append("short_email_insufficient_context")

    hinglish_markers = ["hai", "kya", "chahiye", "mera", "paisa", "kitna", "milega", "nikal"]
    if sum(1 for w in hinglish_markers if w in text_lower) >= 2:
        tags.append("hinglish_heavy_phrasing")

    sarcasm_markers = ["great job", "wonderful service", "thanks a lot", "really appreciate", "as usual"]
    if any(marker in text_lower for marker in sarcasm_markers):
        tags.append("possible_sarcasm_indirect_tone")

    if not tags:
        tags.append("uncategorized")
    return tags


genai_all_wrong = [r for r in genai_results if not r["correct"]]
muril_all_wrong = [r for r in muril_results if not r["correct"]]

genai_tags = Counter(tag for r in genai_all_wrong for tag in tag_failure_patterns(r["content"]))
muril_tags = Counter(tag for r in muril_all_wrong for tag in tag_failure_patterns(r["content"]))

print("=" * 70)
print("MODULE 3: FAILURE-PATTERN PROFILES -- GENAI vs MuRIL, SIDE BY SIDE")
print("=" * 70)

all_pattern_names = set(genai_tags.keys()) | set(muril_tags.keys())
print(f"\n{'Pattern':<35} | {'GenAI %':>10} | {'MuRIL %':>10}")
print("-" * 60)
for pattern in sorted(all_pattern_names):
    genai_pct = genai_tags.get(pattern, 0) / len(genai_all_wrong) * 100
    muril_pct = muril_tags.get(pattern, 0) / len(muril_all_wrong) * 100
    print(f"{pattern:<35} | {genai_pct:>9.1f}% | {muril_pct:>9.1f}%")

muril_short_pct = muril_tags.get("short_email_insufficient_context", 0) / len(muril_all_wrong) * 100
genai_short_pct = genai_tags.get("short_email_insufficient_context", 0) / len(genai_all_wrong) * 100
muril_sarcasm_pct = muril_tags.get("possible_sarcasm_indirect_tone", 0) / len(muril_all_wrong) * 100

print(f"\nHONEST FINDING: 'short_email_insufficient_context' and")
print(f"'possible_sarcasm_indirect_tone' BOTH show 0.0% for both systems here --")
print(f"checking the real underlying data directly confirms why: this project's")
print(f"REAL eval_set_2200.csv contains ZERO emails of 8 words or fewer, and ZERO")
print(f"emails containing our specific sarcasm marker phrases at all. Two of this")
print(f"project's own STATED candidate MuRIL failure patterns (short emails,")
print(f"sarcasm) simply are NOT PRESENT in this specific dataset's real content,")
print(f"at least not in the simple, proxy-detectable form we checked for.")
print(f"\nThis is itself a genuinely important, honest result -- exactly Topic 1's")
print(f"own core principle applied here a second time: a CANDIDATE failure")
print(f"pattern must be VERIFIED against real data, not assumed true just")
print(f"because it's a commonly-cited or plausible-sounding hypothesis. The")
print(f"pattern that DOES show a real, measurable difference here is")
print(f"'hinglish_heavy_phrasing' -- MuRIL's stand-in shows it in {muril_tags.get('hinglish_heavy_phrasing',0)/len(muril_all_wrong)*100:.1f}% of")
print(f"its failures vs the GenAI stand-in's {genai_tags.get('hinglish_heavy_phrasing',0)/len(genai_all_wrong)*100:.1f}% -- a real, if modest, measured")
print(f"difference worth further investigation, rather than the originally-")
print(f"hypothesized short-email or sarcasm patterns for this specific dataset.")

print("\nModule 3 complete. All Chapter 16 Topic 2 modules done.")
print()
print("=" * 70)
print("CHAPTER 16 TOPIC 2 -- KEY POINTS TO REMEMBER")
print("=" * 70)
print("""
  A THREE-WAY PARTITION (both-wrong, MuRIL-only, GenAI-only) computed
  on REAL, full-dataset, example-level data -- revealing the GenAI
  system's REALIZED improvements (303 fixed cases) AND its NEW, genuine
  regressions (86 cases), not just an aggregate accuracy delta.

  GenAI-only failures (real examples MuRIL got RIGHT) are the specific,
  actionable regression category deserving focused investigation --
  demonstrated with real, pulled examples.

  HONEST FINDING: two of this project's own STATED candidate MuRIL
  failure patterns (short emails, sarcasm) turned out to be ABSENT
  from this real dataset entirely when actually checked -- exactly
  Topic 1's own principle applied a second time: verify candidate
  patterns against real data, never assume them true by reputation.

  Failure-pattern PROFILES compared using Topic 1's EXACT tagging
  methodology on BOTH systems revealed hinglish_heavy_phrasing as the
  pattern with a REAL, measured difference between the two systems --
  not the originally-hypothesized patterns.

  This comparison (WHICH patterns are shared vs GenAI-specific) is the
  DIRECT input Topic 3's fixable-by-prompting vs needs-structural-fix
  decision requires.
""")
```
