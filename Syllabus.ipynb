## GenAI End to End with Real World Example of Email Classification  

It provides a practical, hands-on journey from traditional Natural Language Processing (NLP) to advanced Generative AI using a real-world email classification use case.  
Each chapter builds on the previous one, covering the complete pipeline from data preprocessing to transformer models, Large Language Models (LLMs), prompt engineering, Retrieval-Augmented Generation (RAG) and production deployment.

---


### Dataset Reference
- Dataset: `fd_dataset_messy.csv` 
- Distribution: Total 2,500 emails divided into FD samples: 1002, Multiple Category samples: 750 and Non-FD samples: 748
- Source: Claude Synthetic Data
