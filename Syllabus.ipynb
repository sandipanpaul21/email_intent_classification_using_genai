# GenAI Email Classification — Chapter-Wise Study Syllabus

Anchor project: Dataset: `fd_dataset_messy.csv` (2,500 emails — FD 1,002 / Multiple Category 750 / Non-FD 748). Every chapter builds toward a GenAI/agentic layer on top of that baseline.

## Prerequisites
### Chapter 0:  — NLP Foundations
- Sub-topic 1: Tokenization — Word, Character, and Sub-word tokenization; BPE, WordPiece, and SentencePiece compared Refer to `Chapter 0 - 01_Tokenization.ipynb`
- Sub-topic 2: Text representation — One-Hot Encoding, Bag of Words, TF-IDF, and Sparse vs. Dense vectors Refer to  `Chapter 0 - 02_Text_Representation.ipynb`
- Sub-topic 3: Word2Vec — CBOW, Skip-Gram, and Average Word2Vec for sentence-level vectors Refer to `Chapter 0 - 03_Word_Embedding.ipynb`
- Sub-topic 4: Sequence models — RNN, LSTM, GRU, and Bi-LSTM Refer to `Chapter 0 - 04_Embedding_Continue.ipynb`
- Sub-topic 5: Transformer Encoder & BERT Refer to `Chapter 0 - 05_Transformer_Encoder.ipynb`
- Sub-topic 6: Transformer Decoder & GPT Refer to `Chapter 0 - 06_Decoder.ipynb`

### Chapter 1: Calling the Anthropic Model — Basics & Project Setup
- Sub-topic 1: Creating an Anthropic API key (console.anthropic.com) and understanding billing/usage dashboards. Storing keys safely — `.env` files, never hardcoding secrets. Refer to `Chapter 1 - 1_Claude_Model_Comparision.ipynb`
- Sub-topic 2: Loading and inspecting `fd_dataset_messy.csv` — shape, dtypes, nulls, class distribution. Refer to `Chapter 1 - 2_EDA.ipynb`
- Sub-topic 3: Building a stratified dev subset (~300 emails) vs. reserving the full 2,500 for formal evaluation - Refer to `Chapter 1 - 3_Stratified_dev_subset.ipynb`
- Sub-topic 4: Anatomy of a Messages API call — `model`, `messages`, `system`, `max_tokens`, content blocks, `stop_reason`, `usage`, making your first call — sending one email, getting a classification back. Why HAIKU 4.5 because it is POC. Refer to `Chapter 1 - 4_First_API_Call.ipynb`
- Sub-topic 5: Temperature, `top_p`, `max_tokens` — what each controls and why it matters for consistent classification and  Streaming vs. non-streaming responses — when each is appropriate. Refer to `5_Prompt_Output_Filter.ipynb`

### Chapter 2: Tokenization, Prompt Engineering Fundamentals and Prompt Engineering Fundamentals
- Sub-topic 1: Understanding LLM Tokenization: Counting Tokens, Visualizing Hinglish Token Boundaries, Comparing Input vs. Output Tokens, and Calculating Real API Costs. Refer to `Chapter 2 - 1_Tokenization.ipynb`
- Sub-topic 2: Counting tokens before sending a request (token-counting utility)
- Sub-topic 3: Visualizing tokenization on a real Hinglish email from your dataset — seeing the actual token boundaries
- Sub-topic 4: Input tokens vs. output tokens — why they're priced differently
- Sub-topic 5: How code-mixed/Hinglish text tokenizes differently than clean English (the LLM-tokenizer equivalent of your MuRIL UNK-rate findings)
- Sub-topic 6: Calculating real per-call cost from token counts — reconstructing the $25–40 project estimate from first principles

### Chapter 4: Prompt Engineering Fundamentals
- Sub-topic 1: System prompts vs. user messages — what belongs where
- Sub-topic 2: Zero-shot classification — your first working classifier end to end
- Sub-topic 3: Injecting domain knowledge — the 8 keyword groups, negation list, FD reference pattern, class definitions
- Sub-topic 4: Few-shot prompting — adding examples, and how many is actually enough

- Sub-topic 6: Structured output — forcing valid JSON, parsing safely, handling malformed responses
- Sub-topic 7: Prompt versioning — tracking what changed between iterations and why

### Chapter 5: Evaluating a Prompt-Based Classifier
- Sub-topic 1: Building a labeled eval loop over the stratified dev subset
- Sub-topic 2: Computing precision, recall, and F1 for the FD class
- Sub-topic 3: Building a confusion matrix for GenAI predictions vs. ground truth
